```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD
import matplotlib.pyplot as plt
import pandas as pd
```

Importeer het sequentiÃ«le model uit keras.\
Importeer de dense layer uit keras.\
Importeer Adam en SGD uit de keras optimizers\
Importeer matplotlib onder de naam plt\
Importeer pandas onder de naam pd

```python
model = Sequential()
model.add(Dense(1, input_shape=(1,)))
model.summary()
model.compile(Adam(learning_rate=0.8), 'mean_squared_error')
```

Het gaat hier om een sequentieel model voor Lineaire Regressie.
Het model bestaat uit een dense layer met 1 neuron en een input shape layer die dezelfde vorm heeft als de trainingsdata. Door een input_shape mee te geven kan de summary van het model worden getoond.
Verder heeft het model een learning_rate van 0.8. Dit bepaald hoeveel de weight wordt geupdate tijdens het trainen van het model.

#### Adam
Adam optimization is a stochastic gradient descent method that is based on adaptive estimation of first-order and second-order moments.
According to Kingma et al., 2014, the method is "computationally efficient, has little memory requirement, invariant to diagonal rescaling of gradients, and is well suited for problems that are large in terms of data/parameters"

```python
#import seaborn as sns
df = pd.read_csv('datasets/auto_mpg.csv')
df = df[df.horsepower != '?']
df[["mpg", "horsepower"]] = df[["mpg", "horsepower"]].apply(pd.to_numeric)

X = df[['mpg']].values
y_true = df['horsepower'].values
model.fit(X, y_true, epochs=40, verbose=0);
y_pred = model.predict(X)
df.plot(kind='scatter',
        x='mpg',
        y='horsepower',
        title='Horsepower and Miles per Gallon');

plt.plot(X, y_pred, color='red');
W, B = model.get_weights();
print(W)
print(B)
```
